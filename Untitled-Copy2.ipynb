# Splitting of dataset

This data was extracted from the 1994 Census bureau database by Ronny Kohavi and Barry Becker (Data Mining and Visualization, Silicon Graphics). The prediction task is to determine whether a person makes over $50K a year. Some algorithms that could be used for this task include logistic regression, k-nearest neighbors, decision trees, and random forests.

```python
X = data.drop('income',axis=1) 
#target
Y = data['income']

# Split the data into training and testing sets using scikit-learn's train_test_split function. 
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Feature scaling the data using scikit-learn's StandardScaler class.
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Try out different classification algorithms and evaluate their performance using cross-validation
models = {
    "Logistic Regression": LogisticRegression(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Support Vector Machine": SVC(),
    "Neural Network": MLPClassifier()
}

for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv=5)*100
    print(f"{name} Accuracy: {np.mean(scores):.2f}%")
```

# Test your model by making predictions.
# Training and evaluating
Now that we have our data preprocessed and split into training and test sets, we can begin building predictive models. We'll start by training a **Gradient Boosting Classifier** model and evaluating its performance on the test set.

### Gradient Boosting Classifier

```python
# Build and evaluate a Gradient Boosting Classifier model
gbc = GradientBoostingClassifier(random_state=42)
gbc.fit(X_train, y_train)
y_pred = gbc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Gradient Boosting Classifier: {accuracy:.4f}')
```

I get an accuracy score of ~0.87, which indicates that the Gradient Boosting Classifier model is performing well on the test set.

```python
# Evaluate the models
print("Gradient Boosting Classifier")
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))

# Compute cross-entropy loss
y_proba = gbc.predict_proba(X_test)
cross_entropy = log_loss(y_test, y_proba)

print("Cross-Entropy Loss:", cross_entropy)
```

The log_loss function is used to compute the cross-entropy loss for the test set.

#### Cross entropy loss (or logic loss)

A large difference will contain a huge score such as 0.9 or 10. Smaller differences will contain small points such as 0.1 or 0.2.

A model with exact probability will contain a logic loss or cross entropy of 0.0. This indicates that the cross-entropy loss is minimal, and smaller values will represent a good model rather than larger ones.

### Evaluation of a binary classification model using precision-recall curves and F1 score

```python
# Compute precision-recall curve and F1 score
y_proba = gbc.predict_proba(X_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test, y_proba)

f1_scores = []
for i in range(len(thresholds)):
    y_pred = (y_proba > thresholds[i]).astype(int)
    f1_scores.append(f1_score(y_test, y_pred))
    
plt.plot(recall, precision, label='Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower right')

plt.show()

plt.plot(thresholds, f1_scores, label='F1 Score')
plt.xlabel('Threshold')
plt.ylabel('F1 Score')
plt.title('F1 Score vs. Threshold')
plt.legend(loc='lower right')

plt.show()
```

Model **recall score** represents the model's ability to **correctly predict the positives out of actual positives**. This is unlike precision which measures how many predictions made by models are actually positive out of all positive predictions made.

The highest possible value of an F-score is 1.0, indicating perfect precision and recall, and the lowest possible value is 0, if either precision or recall are zero. My model achieves a precision of approximately **0.7 f1 factor** for a **threshold of 0.4**.

### Random Forest

Next, we'll try an ensemble method, a Random Forest. We'll use scikit-learn's Random Forest class to train the Random Forest, and again evaluate its performance on the test set.

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Random forest accuracy: {accuracy:.3f}')
```

I get an accuracy score of 0.860, which is bit lower value then for Gradient Boosting Classifier. This suggests that either the Gradient Boosting Classifier model or the Random Forest model could be a good choice for this particular problem.

```python
print("Random Forest")
print(classification_report(y_test, rf.predict(X_test)))
print(confusion_matrix(y_test, rf.predict(X_test)))
```

### Neural Network Accuracy

Finally, we'll try a different classification algorithm, a Neural Network Accuracy model. We'll use scikit-learn's RandomForestClassifier class to train the random forest, and once again evaluate its performance on the test set.

```python
clf = MLPClassifier(random_state=42)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Neural Network accuracy: {accuracy:.3f}')
```

I get an accuracy score of 0.857, which is slightly lower than the Gradient Boosting Classifier model. This suggests that the Gradient Boosting Classifier model may be a better choice for this particular problem.

```python
# Evaluate the models
print("Support Vector Machine")
print(classification_report(y_test, clf.predict(X_test)))
print(confusion_matrix(y_test, clf.predict(X_test)))
```

### Evaluate models using CV

```python
# Evaluate models using cross-validation
models = [gbc, rf, clf]
model_names = ['Gradient Boosting Classifier', 'Random Forest', 'Neural Network Accuracy']

for i, model in enumerate(models):
    scores = cross_val_score(model, X, Y, cv=5, scoring='accuracy')
    print(model_names[i])
    print('Cross-validated scores:', scores)
    print('Mean cross-validated score:', np.mean(scores))
    
    y_proba = model.predict_proba(X_test)[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, y_proba)
    auc_score = roc_auc_score(y_test, y_proba)
    
    plt.plot(fpr, tpr, label='AUC = %0.2f' % auc_score)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc='lower right')

plt.show()
```

**In this example**, we evaluated the models using cross-validation. For each model, we computed the cross-validated accuracy scores using cross_val_score() and reported the mean score. We also computed the AUC score and plotted the ROC curve for each model using roc_curve() and roc.
